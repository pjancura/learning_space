```python
# imports
import plotly
from plotly import __version__
print(__version__)
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import plotly.graph_objs as go
import chart_studio.plotly as cs
import chart_studio.tools as cstools
import pandas as pd

# Jupyter setup
init_notebook_mode(connected=True)
```

```python
tesla = pd.read_csv('P48-Time-Series-Tesla-Stock-Price.csv')
```

```python
tesla.head(10)
```

```python
trace_1 = go.Scatter(
            x = tesla.date,
            y = tesla['high'],
            name = 'Tesla High',
            line = dict(
                color = '#17BECF'),
                opacity = 0.8)

trace_2 = go.Scatter(
            x = tesla.date,
            y = tesla['low'],
            name = 'Tesla Low',
            line = dict(
                color = '#7F7F7F'),
                opacity = 0.8)

data = [trace_1, trace_2]

layout = dict(
            title = 'Tesla Stock Price - High vs Low')

fig = dict(data = data, layout = layout)
iplot(fig, filename='tesla_stock_comparison')
```
